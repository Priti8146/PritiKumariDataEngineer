```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, collect_set

spark = SparkSession.builder.appName("EmployeesInBothDepts").getOrCreate()

# Sample data: emp_id, name, department_id
data = [
    (1, "Alice", 101),
    (1, "Alice", 102),
    (2, "Bob", 101),
    (3, "Charlie", 102),
    (3, "Charlie", 103),
    (4, "David", 101),
    (4, "David", 102),
    (5, "Eve", 104)
]

columns = ["emp_id", "name", "department_id"]

df = spark.createDataFrame(data, columns)
df.show()
```

How to find all employees who work in both departments 101 and 102?


```python
from pyspark.sql import functions as F

# Group by emp_id and name, aggregate department_ids
emp_depts = df.groupBy("emp_id", "name").agg(F.collect_set("department_id").alias("departments"))
emp_depts.display()
```
